# Random Walk Graph Plotter

This notebook allows the user to plot network characteristics for a given number of nodes, a min and max time step, and random seed value.

To use:
`graph(timeMin, timeMax, numNodes, seed)` where

`timeMin` is the minimum time value to be plotted along the x-axis (at least one)

`timeMax` is the maximum time value to be iterated through

`numNodes` is the number of agents/nodes (for now, must be a perfect square)

`seed` is any integer for the random number generator (for testing purposes)

Returns:
Two plots: mean path length and clustering coefficient for the same graph given a number of time steps, and the average degree vs time. (There is also the possibility of returning a degree distribution histogram in `networkx` but it is difficult to implement given many time steps.)

**Notes on what you're looking at** According to Muthukrishna and Schaller [(link here)](https://doi-org.libproxy.boisestate.edu/10.1177%2F1088868319855783), realistic social networks can be categorized with the following characteristics:

1. Degree distribution that is skewed to the right (to represent that most individuals have at least a few acquaintances, but few individuals have extremely large numbers of acquaintances.) *This is indicated by a positive skewness value.*


2. Average clustering coefficient. From Muthukrishna and Schaller: 
> The second property refers to the likelihood that any two
acquaintances of any individual will also be acquainted
with each other. Within real social networks, this likelihood
is **non-zero**, which is reflected in indices that assess the
“clustering” of social connections within the network.
> 


3. Average path length. Again, Muthukrishna and Schaller: 
> The
third property refers to the average smallest number of
social connections required to trace a path from any one
individual to any other individual within the population
(the “average path length”). For instance, the online social
network Facebook—a means of communication and social
influence in contemporary populations—has an average
path length between **3 and 4** (Backstrom, Boldi, Rosa,
Ugander, & Vigna, 2012; Boldi & Vigna, 2012).
>
